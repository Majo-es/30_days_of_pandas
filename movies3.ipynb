### FILTER ROWS:

```python
booleans = []
for lenght in movies.duration:
    if lenght >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
# SAME LENGHT AS THE DATAFRAME
len(booleans)
```

```python
# CONVERT LIST INTO A PANDAS SERIES
long_list = pd.Series(booleans)
```

```python
long_list.head()
```

```python
long_list = movies.duration >= 200
long_list.head()
```

```python
#  filtering the movies DataFrame
movies[movies.duration >= 200]
```

```python
# PASS THE PANDAS SERIES USING BRACKET CONVERSION
movies[long_list]
```

```python
# returns a pandas Series
movies['genre']
```

```python
#  filtering the movies DataFrame
movies[movies.duration >= 200].genre
```

```python
# USING BRACKET NOTATION
movies[movies.duration >= 200]['genre']
```

```python
#  filter the movies DataFrame and select a specific column using the .loc accessor.
movies.loc[movies.duration >= 200, 'genre']
```
