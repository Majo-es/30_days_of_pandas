### Sorting a series:

````python
# SORT USING DOT NOTATION: 1
# ALPHABETICAL OLDER. NUMBERS COME FIRST
movies.title.sort_values()
````

````python
# SORT USING BRACKET NOTATION: 2
# ALPHABETICAL OLDER. NUMBERS COME FIRST
movies['title'].sort_values()
````

````python
movies['title'].sort_values(ascending=False)
````

````python
type(movies['title'].sort_values())
````

````python
# ORIGINAL ORDER
movies['title']
````

### Sorting a dataframe by a series

````python
movies.sort_values('title')
````

````python
movies.sort_values('duration')
````

````python
movies.sort_values('duration', ascending=False)
````

````python
movies.sort_values(['star_rating','title'],ascending=[False,True])
````
