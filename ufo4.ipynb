```python
ufo = pd.read_csv('http://bit.ly/uforeports', usecols= ['City', 'State'])
ufo.columns
```

```python
# usecols argument is used to specify which columns to read from the CSV file
ufo = pd.read_csv('http://bit.ly/uforeports', usecols= [0,3])
ufo.columns
```

```python
# The nrows=3 argument tells pd.read_csv() to only read the first 3 rows of the CSV file (in addition to the header row)
ufo = pd.read_csv('http://bit.ly/uforeports', nrows=3)
ufo
```

### iterate through a Series or a DataFrame

```python
# for loop that iterates over each item in the ufo.City Series
for c in ufo.City:
    print(c)
```

```python
#This for loop unpacks each pair generated by iterrows(). In each iteration:
#index gets the index of the current row.
#row gets a pandas Series containing the data for the current row. The Series is indexed by the column names.
for index, row in ufo.iterrows():
    print(index, row.City, row.State)
```
