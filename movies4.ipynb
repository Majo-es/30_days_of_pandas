### APPLY MULTIPLE FILTER CRITERIA TO A PANDAS DATAFRAME:

````python
movies[movies.duration >= 200]
````

#### Or Operator
```
True or False
> True
```

```
True or True
> True
```

```
False or False
> False
```

#### And Operator

```
True and False
> False
```

```
True and True
> True
```







````python
# AND
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
````

````python
(movies.duration >= 200) & (movies.genre == 'Drama')
````

````python
# OR
movies[(movies.duration >= 200) | (movies.genre == 'Drama')]
````

````python
(movies.duration >= 200) | (movies.genre == 'Drama')
````

````python
movies[(movies.duration == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action')]
````

````python
# SIMPLIFY:
# SERIES METHOD & PASS A LIST -> BOOLEAN SERIES
movies.genre.isin(['Crime', 'Drama', 'Action'])
````

````python
# PASS THE BOOLEAN SERIES USING BRACKET NOTATION TO THE DATAFRAME
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])]
````
