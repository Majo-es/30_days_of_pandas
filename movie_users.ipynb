```python
# READ TABULAR DATA (ROWS & COLUMNS):
import pandas as pd

# Use pd.read_table to read the tabular data
pd.read_table('http://bit.ly/movieusers' , sep='|', header=None)
```

```python
# LIST OF STRINGS
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']

# DATAFRAME
user = pd.read_table('http://bit.ly/movieusers' , sep='|', header=None, names= user_cols)
```

```python
user.head(10)
```

```python
def format_color_groups(user):
  # LIST
    colors = ['gray', 'purple']
    x = user.copy()
    factors = list(x['gender'].unique())
    # INDEX
    i = 0
    # ITERATE
    for factor in factors:
        # .STYLE: FORMATTING
        style = f'background-color: {colors[i]}'
        # .LOC: SELECT ALL ROWS WHERE THE "GENDER" MATCHES CURRENT FACTOR
        x.loc[x['gender'] == factor, :] = style
        i = not i
    return x

# CALL IT
user.style.apply(format_color_groups, axis=None)
```
